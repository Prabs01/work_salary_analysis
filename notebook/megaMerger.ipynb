```python
import pandas as pd
import numpy as np
```

```python
#lets create a mega file merging all the datasets needed for radar plot for more flexibility
df1 = pd.read_csv('../data/2018/Earning2018.csv',header= None)
df2 = pd.read_csv('../data/2018/employePopn2018.csv',header= None)
df3 = pd.read_csv('../data/2018/workinghr2018.csv',header= None)

df4 = pd.read_csv('../data/2008/Earning2008.csv',header= None)
df5 = pd.read_csv('../data/2008/employePopn2008.csv',header= None)
df6 = pd.read_csv('../data/2008/workinghr2008.csv',header= None)

df7 = pd.read_csv('../data/1998/Earning1998.csv',header= None)
df8 = pd.read_csv('../data/1998/employePopn1998.csv',header= None)
df9 = pd.read_csv('../data/1998/workinghr1998.csv',header= None)
```

```python
df1.columns = df1.iloc[0]
df1 = df1.drop(0).reset_index(drop=True)
df1 = df1.rename(columns={df1.columns[0]: "Occupation"})  # Ensure first column is named "Occupation"

# Repeat for df2 and df3
df2.columns = df2.iloc[0]
df2 = df2.drop(0).reset_index(drop=True)
df2 = df2.rename(columns={df2.columns[0]: "Occupation"})

df3.columns = df3.iloc[0]
df3 = df3.drop(0).reset_index(drop=True)
df3 = df3.rename(columns={df3.columns[0]: "Occupation"})


df4.columns = df4.iloc[0]
df4 = df4.drop(0).reset_index(drop=True)
df4 = df4.rename(columns={df4.columns[0]: "Occupation"})

df5.columns = df5.iloc[0]
df5 = df5.drop(0).reset_index(drop=True)
df5 = df5.rename(columns={df5.columns[0]: "Occupation"})

df6.columns = df6.iloc[0]
df6 = df6.drop(0).reset_index(drop=True)
df6 = df6.rename(columns={df6.columns[0]: "Occupation"})


df7.columns = df7.iloc[0]
df7 = df7.drop(0).reset_index(drop=True)
df7 = df7.rename(columns={df7.columns[0]: "Occupation"})

df8.columns = df8.iloc[0]
df8 = df8.drop(0).reset_index(drop=True)
df8 = df8.rename(columns={df8.columns[0]: "Occupation"})

df9.columns = df9.iloc[0]
df9 = df9.drop(0).reset_index(drop=True)
df9 = df9.rename(columns={df9.columns[0]: "Occupation"})
```

```python
# Merge of respective years first
merged_df_2018 = df1.merge(df2, on="Occupation", how="outer") \
               .merge(df3, on="Occupation", how="outer")
#rename the columns for ease
merged_df_2018.columns = ['Occupation', 'total_ear_2018', 'male_ear_2018', 'female_ear_2018',
                        'total_pop_2018', 'male_pop_2018', 'female_pop_2018',
                        'total_wor_2018', 'male_wor_2018', 'female_wor_2018']
```

```python
merged_df_2018
```

```python
# Merge of respective years first
merged_df_2008 = df4.merge(df5, on="Occupation", how="outer") \
               .merge(df6, on="Occupation", how="outer")
#rename the columns for ease
merged_df_2008.columns = ['Occupation', 'total_ear_2008', 'male_ear_2008', 'female_ear_2008',
                        'total_pop_2008', 'male_pop_2008', 'female_pop_2008',
                        'total_wor_2008', 'male_wor_2008', 'female_wor_2008']
```

```python
# Merge of respective years first
merged_df_1998 = df7.merge(df8, on="Occupation", how="outer") \
               .merge(df9, on="Occupation", how="outer")
#rename the columns for ease
merged_df_1998.columns = ['Occupation', 'total_ear_1998', 'male_ear_1998', 'female_ear_1998',
                        'total_pop_1998', 'male_pop_1998', 'female_pop_1998',
                        'total_wor_1998', 'male_wor_1998', 'female_wor_1998']
```

```python
#lerts merge all
merged_df = merged_df_2018.merge(merged_df_2008, on="Occupation", how="outer") \
               .merge(merged_df_1998, on="Occupation", how="outer")
```

```python
# Assuming 'merged_df' is your final DataFrame
merged_df.to_csv('..\data\megaMergedData.csv', index=False)  # Save to 'output.csv' without row index
```
